## Consistency and contribution metrics
Here I calculate average runs, not outs, average runs per innings and each player's contribution to the season.

```python
#avg_runs=df.groupby("Player",as_index=False)["Runs"].mean()
#print(avg_runs)  #return as dataframe  dataframe have both columns names and index values in o/p 
```

```python
avg_runs = df.groupby("Player")["Runs"].mean()  #return as series  series dont have index and column name in o/p
avg_runs=avg_runs.sort_values(ascending=False).head(11)
print(avg_runs)
```

```python
not_outs=df.sort_values(by="NO",ascending=False)[["Player","Inns","NO"]].head(11)
print(not_outs)
```

```python
df["Avg_runs_per_innings"] = df["Runs"] / df["Inns"]
print(df[["Player","Avg_runs_per_innings"]])
```

```python
total_runs = df["Runs"].sum()
df["Season_Contribution_%"] = df["Runs"] / total_runs * 100
print(df[["Player", "Season_Contribution_%"]])
```

## Boundary behaviour (4s and 6s)
In this part I check how many 4s and 6s players hit on average per innings.

```python
df["avg6s"]=df["6s"] / df["Inns"]
avg6s=df.sort_values(by="avg6s",ascending=False)[["Player","Inns","avg6s"]].head(11)
print(avg6s)
```

```python
df["avg4s"]=df["4s"] / df["Inns"]
avg4s=df.sort_values(by="avg4s",ascending=False)[["Player","Inns","avg4s"]].head(11)
print(avg4s)
```

## Top scorers and extra metrics
Now I summarise some top scorers and then create a few extra helpful metrics like boundary runs, boundary percentage and runs per ball.

```python

# Top 10 run scorers
top_runs = df.sort_values(by="Runs", ascending=False).head(10)
top_runs[["Player", "Runs", "BF", "SR", "4s", "6s"]]
```

```python

# Create a few helpful metrics

# Boundary runs = 4s*4 + 6s*6
df["Boundary_Runs"] = df["4s"] * 4 + df["6s"] * 6

# Boundary percentage of total runs
df["Boundary_%"] = (df["Boundary_Runs"] / df["Runs"]) * 100

# Runs per ball (scoring efficiency)
df["Runs_per_Ball"] = df["Runs"] / df["BF"]

df[["Player", "Runs", "BF", "Runs_per_Ball", "Boundary_%"]].head()
```

```python

# Top 10 most efficient scorers (min 100 balls faced to avoid tiny samples)
efficient = df[df["BF"] >= 100].sort_values(by="Runs_per_Ball", ascending=False).head(10)
efficient[["Player", "Runs", "BF", "Runs_per_Ball", "Boundary_%", "SR"]]
```

## Visual exploration
Finally I plot a few charts to visualise relationships and distributions in the data.

```python

# Runs vs Balls Faced
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="BF", y="Runs")
plt.title("Runs vs Balls Faced")
plt.xlabel("Balls Faced")
plt.ylabel("Runs")
plt.tight_layout()
plt.show()
```

```python

# Strike Rate vs Boundary % (for players with minimum balls faced)
filtered = df[df["BF"] >= 100]

plt.figure(figsize=(10, 6))
sns.scatterplot(data=filtered, x="Boundary_%", y="SR")
plt.title("Strike Rate vs Boundary % (BF â‰¥ 100)")
plt.xlabel("Boundary % of Runs")
plt.ylabel("Strike Rate")
plt.tight_layout()
plt.show()
```

```python

# Distribution of Runs
plt.figure(figsize=(10, 5))
sns.histplot(df["Runs"], bins=20, kde=True)
plt.title("Distribution of Runs")
plt.xlabel("Runs")
plt.ylabel("Count of Players")
plt.tight_layout()
plt.show()
```

```python

# Distribution of Strike Rate
plt.figure(figsize=(10, 5))
sns.histplot(df["SR"], bins=20, kde=True)
plt.title("Distribution of Strike Rate")
plt.xlabel("Strike Rate")
plt.ylabel("Count of Players")
plt.tight_layout()
plt.show()
```

```python

# Select numeric columns for correlation
numeric_cols = df.select_dtypes(include=[np.number])
corr = numeric_cols.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

## My quick insights and notes

Some quick observations from this analysis (from my point of view):

- Some players contribute a big percentage of the total season runs for their team.
- Average runs per innings is more useful for judging consistency than just total runs.
- Boundary percentage has a clear relation with strike rate: players with higher boundary % usually have higher SR.
- Players who score quickly with fewer balls faced stand out as very efficient scorers.
